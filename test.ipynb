```python
import fitz  # PyMuPDF
from docx import Document

def extract_text_and_tables(pdf_path):
    doc = fitz.open(pdf_path)
    content = []
    
    for page_num in range(len(doc)):
        page = doc[page_num]
        text = page.get_text("text")  # Extract plain text
        blocks = page.get_text("dict")["blocks"]  # Extract blocks (text and tables)
        
        page_content = {"page_num": page_num + 1, "text": text, "tables": []}

        # Process blocks to identify potential tables
        for block in blocks:
            if "lines" in block:  # Only process text-containing blocks
                table_data = []
                for line in block["lines"]:
                    row = " ".join([span["text"] for span in line["spans"]])
                    table_data.append(row)
                if len(table_data) > 1:  # Potential table
                    page_content["tables"].append(table_data)

        content.append(page_content)
    
    return content

# Example usage
pdf_path = "Voterslist.pdf"
extracted_content = extract_text_and_tables(pdf_path)
print(extracted_content)
```

```python
from pypdf import PdfReader
import fitz  # PyMuPDF
import easyocr

def extract_text_from_page(pdf_path, page_number):
    # Initialize readers
    pdf_reader = PdfReader(pdf_path)
    ocr_reader = easyocr.Reader(['hi'])  # Add desired languages, e.g., ['en', 'hin']

    # Ensure page_number is valid
    if page_number < 1 or page_number > len(pdf_reader.pages):
        raise ValueError(f"Invalid page number: {page_number}. PDF has {len(pdf_reader.pages)} pages.")
    
    # Extract text using pypdf
    page = pdf_reader.pages[page_number - 1]
    text = page.extract_text()

    if text:
        return f"--- Page {page_number} (Text Extracted) ---\n{text}\n"
    else:
        # If text is not directly extractable, use OCR
        pdf_doc = fitz.open(pdf_path)
        pdf_page = pdf_doc[page_number - 1]

        # Render page as an image
        pix = pdf_page.get_pixmap(dpi=300)  # Render at 300 DPI for better OCR results
        image_data = pix.tobytes("png")  # Convert to PNG byte format

        # Use EasyOCR to extract text from the image
        ocr_results = ocr_reader.readtext(image_data, detail=0, paragraph=True)
        ocr_text = "\n".join(ocr_results)
        
        return f"--- Page {page_number} (OCR Extracted) ---\n{ocr_text}\n"

# Example usage
pdf_path = "Voterslist.pdf"
page_number = 1  # Specify the page number to process
try:
    result = extract_text_from_page(pdf_path, page_number)
    print(result)

    # Optionally save to a file
    with open(f"output_page_{page_number}.txt", "w", encoding="utf-8") as f:
        f.write(result)
except ValueError as e:
    print(e)
```

```python
import os
import easyocr
import numpy as np
from pdf2image import convert_from_path
import docx
from PIL import Image

def convert_pdf_to_text_and_docx(pdf_path, output_text_path=None, output_docx_path=None, languages=['hi', 'en']):
    """
    Convert a PDF to text and DOCX using EasyOCR
    
    Args:
    pdf_path (str): Path to the input PDF file
    output_text_path (str, optional): Path to save the extracted text file
    output_docx_path (str, optional): Path to save the DOCX file
    languages (list, optional): List of languages to recognize
    
    Returns:
    str: Extracted text from the PDF
    """
    # Convert PDF to images
    images = convert_from_path(pdf_path)
    
    # Initialize EasyOCR reader
    reader = easyocr.Reader(languages)
    
    # Extract text from images
    full_text = ""
    for i, image in enumerate(images):
        # Convert PIL Image to numpy array
        image_np = np.array(image)
        
        # Perform OCR
        results = reader.readtext(image_np)
        
        # Extract text from results
        page_text = " ".join([result[1] for result in results])
        full_text += page_text + "\n\n"
    
    # Save as plain text if requested
    if output_text_path:
        with open(output_text_path, 'w', encoding='utf-8') as text_file:
            text_file.write(full_text)
    
    # Save as DOCX if requested
    if output_docx_path:
        doc = docx.Document()
        doc.add_paragraph(full_text)
        
        # Set font to support multiple languages
        for paragraph in doc.paragraphs:
            for run in paragraph.runs:
                run.font.name = 'Arial Unicode MS'
        
        doc.save(output_docx_path)
    
    return full_text

def main():
    # Example usage
    pdf_path = 'sample.pdf'
    
    # Convert to text
    text_output = 'extracted_text.txt'
    convert_pdf_to_text_and_docx(
        pdf_path, 
        output_text_path=text_output,
        languages=['hi', 'en']  # Support both Hindi and English
    )
    
    # Convert to DOCX
    docx_output = 'extracted_document.docx'
    convert_pdf_to_text_and_docx(
        pdf_path, 
        output_docx_path=docx_output,
        languages=['hi', 'en']
    )
    
    print(f"Conversion complete. Text saved to {text_output} and DOCX saved to {docx_output}")

if __name__ == "__main__":
    main()
```

```python
def wrap_text(input_text, max_width=100):
    wrapped_text = []
    lines = input_text.split("\n")  # Split into individual lines for processing

    for line in lines:
        if len(line) > max_width:  # Check if the line exceeds max width
            # Wrap the line
            while len(line) > max_width:
                # Split the line and append to wrapped_text
                wrapped_text.append(line[:max_width])
                line = line[max_width:].strip()  # Take the remaining part of the line
            # Append any remaining part of the line
            if line:
                wrapped_text.append(line)
        else:
            wrapped_text.append(line)

    return "\n".join(wrapped_text)


# Example Input
input_text = input_text
# Apply Formatting
formatted_text = wrap_text(input_text, max_width=100)

# Output Result
print(formatted_text)
```

```python
import cv2
import numpy as np

def detect_tables(image_path):
    # Load the image
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

    # Define kernels for morphological operations
    horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 1))
    vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 25))

    # Detect horizontal lines
    horizontal_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

    # Detect vertical lines
    vertical_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, vertical_kernel, iterations=2)

    # Combine horizontal and vertical lines
    table_structure = cv2.add(horizontal_lines, vertical_lines)

    # Find contours in the table structure
    contours, _ = cv2.findContours(table_structure, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw bounding boxes around detected tables
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        if w > 50 and h > 50:  # Filter out small boxes
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Show or save the result
    # cv2.imshow("Detected Tables", image)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()
    cv2.imwrite("output.png", image)

# Example usage
detect_tables("page_1.png")
```

```python
import cv2
import easyocr
import numpy as np

def extract_text_from_first_rectangle(image_path):
    # Load the image
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding (invert binary for better contrast of text)
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

    # Define kernels for morphological operations
    horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 1))
    vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 25))

    # Detect horizontal lines
    horizontal_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

    # Detect vertical lines
    vertical_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, vertical_kernel, iterations=2)

    # Combine horizontal and vertical lines
    table_structure = cv2.add(horizontal_lines, vertical_lines)

    # Find contours in the table structure
    contours, _ = cv2.findContours(table_structure, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Extract the first contour's bounding box (assuming the top-left rectangle is the first contour)
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        if w > 50 and h > 50:  # Filter out small boxes
            # Crop the region of interest (ROI) based on bounding box
            roi = image[y:y+h, x:x+w]

            # Display the cropped region for debugging
            cv2.imshow("Cropped ROI", roi)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

            # Initialize EasyOCR reader
            reader = easyocr.Reader(['hi', 'en'])

            # Extract text from the cropped region
            result = reader.readtext(roi)

            # Print the detected text
            extracted_text = "\n".join([text[1] for text in result])
            print("Extracted Text from First Rectangle:\n", extracted_text)
            break  # Stop after the first rectangle is processed

# Example usage
extract_text_from_first_rectangle("page_1.png")
```

```python
import cv2
import numpy as np
import easyocr

def detect_tables_and_extract_text(image_path, output_txt_path):
    # Initialize EasyOCR reader
    reader = easyocr.Reader(['en','hi'])

    # Load the image
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

    # Define kernels for morphological operations
    horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 1))
    vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 25))

    # Detect horizontal lines
    horizontal_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

    # Detect vertical lines
    vertical_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, vertical_kernel, iterations=2)

    # Combine horizontal and vertical lines
    table_structure = cv2.add(horizontal_lines, vertical_lines)

    # Find contours in the table structure
    contours, _ = cv2.findContours(table_structure, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # List to store extracted text for each table (10 rows, 3 columns)
    table_data = []

    # Draw bounding boxes around detected tables and extract text
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        if w > 50 and h > 50:  # Filter out small boxes
            # Extract text using EasyOCR from the detected bounding box
            table_img = image[y:y + h, x:x + w]
            result = reader.readtext(table_img)

            # Initialize a list to store row-wise text (each row should have 3 columns)
            table_rows = []

            # Loop through the OCR results and organize them into rows
            for i in range(0, len(result), 3):  # Assuming 3 columns per row
                row = [text for _, text, _ in result[i:i+3]]
                table_rows.append(row)

            # Append the rows to the table data
            table_data.append(table_rows)

            # Draw the bounding box around the table
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Save the modified image with bounding boxes
    cv2.imwrite("output_detected_tables.png", image)

    # Write the extracted text into a text file
    with open(output_txt_path, 'w') as f:
        for table in table_data:
            for row in table:
                f.write('\t'.join(row) + '\n')  # Write each row with tab-separated columns
            f.write('\n')  # Add a newline between different tables

# Example usage
detect_tables_and_extract_text("page.png", "extracted_table_data.txt")
```

```python
from pdf2image import convert_from_path
import cv2
import numpy as np
import easyocr
import os

def detect_tables_and_extract_text_from_pdf(pdf_path, output_txt_path, output_image_dir):
    # Initialize EasyOCR reader
    reader = easyocr.Reader(['en', 'hi'])

    # Convert PDF to images
    pages = convert_from_path(pdf_path, dpi=300)
    
    # Ensure the output image directory exists
    os.makedirs(output_image_dir, exist_ok=True)

    # List to store extracted text from all pages
    all_tables_data = []

    # Process each page
    for page_num, page_image in enumerate(pages, start=1):
        # Save page image as a temporary file
        page_image_path = os.path.join(output_image_dir, f"page_{page_num}.png")
        page_image.save(page_image_path, "PNG")

        # Load the image
        image = cv2.imread(page_image_path)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Apply binary thresholding
        _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

        # Define kernels for morphological operations
        horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 1))
        vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 25))

        # Detect horizontal lines
        horizontal_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

        # Detect vertical lines
        vertical_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, vertical_kernel, iterations=2)

        # Combine horizontal and vertical lines
        table_structure = cv2.add(horizontal_lines, vertical_lines)

        # Find contours in the table structure
        contours, _ = cv2.findContours(table_structure, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # List to store extracted text for each table
        page_table_data = []

        # Draw bounding boxes around detected tables and extract text
        for contour in contours:
            x, y, w, h = cv2.boundingRect(contour)
            if w > 50 and h > 50:  # Filter out small boxes
                # Extract text using EasyOCR from the detected bounding box
                table_img = image[y:y + h, x:x + w]
                result = reader.readtext(table_img)

                # Initialize a list to store row-wise text
                table_rows = []

                # Loop through the OCR results and organize them into rows
                for i in range(0, len(result), 3):  # Assuming 3 columns per row
                    row = [text for _, text, _ in result[i:i+3]]
                    table_rows.append(row)

                # Append the rows to the page's table data
                page_table_data.append(table_rows)

                # Draw the bounding box around the table
                cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Save the modified image with bounding boxes
        output_image_path = os.path.join(output_image_dir, f"page_{page_num}_detected.png")
        cv2.imwrite(output_image_path, image)

        # Append the page's table data to the overall list
        all_tables_data.append(page_table_data)

    # Write all extracted text into a single text file
    with open(output_txt_path, 'w') as f:
        for page_num, page_tables in enumerate(all_tables_data, start=1):
            f.write(f"--- Page {page_num} ---\n")
            for table in page_tables:
                for row in table:
                    f.write('\t'.join(row) + '\n')  # Write each row with tab-separated columns
                f.write('\n')  # Add a newline between different tables
            f.write('\n')  # Add a newline between different pages

# Example usage
detect_tables_and_extract_text_from_pdf(
    "sample.pdf",
    "extracted_data.txt",
    "output.png"
)
```

```python
from pdf2image import convert_from_path
import cv2
import numpy as np
import easyocr
import os
from docx import Document

def detect_tables_and_extract_text_to_docx(pdf_path, output_docx_path, output_image_dir):
    # Initialize EasyOCR reader
    reader = easyocr.Reader(['en', 'hi'])

    # Convert PDF to images
    pages = convert_from_path(pdf_path, dpi=300)
    
    # Ensure the output image directory exists
    os.makedirs(output_image_dir, exist_ok=True)

    # Create a Word document
    doc = Document()

    # Process each page
    for page_num, page_image in enumerate(pages, start=1):
        # Add a heading for the page
        doc.add_heading(f"Page {page_num}", level=1)

        # Save page image as a temporary file
        page_image_path = os.path.join(output_image_dir, f"page_{page_num}.png")
        page_image.save(page_image_path, "PNG")

        # Load the image
        image = cv2.imread(page_image_path)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Apply binary thresholding
        _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

        # Define kernels for morphological operations
        horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 1))
        vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 25))

        # Detect horizontal lines
        horizontal_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

        # Detect vertical lines
        vertical_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, vertical_kernel, iterations=2)

        # Combine horizontal and vertical lines
        table_structure = cv2.add(horizontal_lines, vertical_lines)

        # Find contours in the table structure
        contours, _ = cv2.findContours(table_structure, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Draw bounding boxes around detected tables and extract text
        for contour in contours:
            x, y, w, h = cv2.boundingRect(contour)
            if w > 50 and h > 50:  # Filter out small boxes
                # Extract text using EasyOCR from the detected bounding box
                table_img = image[y:y + h, x:x + w]
                result = reader.readtext(table_img)

                # Add a table to the Word document
                table = doc.add_table(rows=0, cols=3)  # Assuming 3 columns per table row

                # Loop through the OCR results and organize them into rows
                for i in range(0, len(result), 3):  # Assuming 3 columns per row
                    row = [text for _, text, _ in result[i:i+3]]
                    doc_row = table.add_row().cells
                    for idx, cell_text in enumerate(row):
                        if idx < len(doc_row):
                            doc_row[idx].text = cell_text

        # Save the modified image with bounding boxes
        output_image_path = os.path.join(output_image_dir, f"page_{page_num}_detected.png")
        cv2.imwrite(output_image_path, image)

    # Save the Word document
    doc.save(output_docx_path)

# Example usage
detect_tables_and_extract_text_to_docx(
    "sample.pdf",
    "extracted_table_data.docx",
    "output.jpg"
)
```
